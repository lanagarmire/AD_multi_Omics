```r
data <- read.csv("extdata/all_lmci_metab_hmdb_lilikoi.csv", header=TRUE)

name_map <- read.csv("extdata/name_map.csv",header=TRUE)
```

```r
data_1 = data[,c(-1,-2)]
```

```r
rownames(data_1) = data$PID #rownames(data)
```

```r
data_2 = t(data_1)
```

```r
dim(data_2)
```

```r
subtype_info<-read.csv("extdata/cn_emci_lmci_ad_clin_pseudotime.csv",header=TRUE)
table(subtype_info$SNF)
name_emci1<-subtype_info$PID[which(subtype_info$SNF=="lmcisubtype1")]
name_emci2<-subtype_info$PID[which(subtype_info$SNF=="lmcisubtype2")]
```

```r
data_subset = data_2[,c(name_emci1,name_emci2)]
# subset clusters for c=2 SNF MRNA + METAB
full_clinical =  read.csv('extdata/cn_emci_lmci_ad_clin_pseudotime.csv',row.names=1)
pd = full_clinical[colnames(data_2),c('SNF','AGE','PTGENDER','PTRACCAT')]
pd$PTGENDER = as.character(pd$PTGENDER)
pd$PTRACCAT = as.character(pd$PTRACCAT)
pd = pd[pd$SNF%in%c('lmcisubtype1','lmcisubtype2'),]
```

```r
formula = paste0(names(pd), collapse = ' + ')
formula = paste0("~", formula)
formula = formula(formula)
design = model.matrix(formula, data = pd)
design[1:5,]
```

```r
fit1 = lmFit(data_2, design)
fit2 = eBayes(fit1)
limma_result_fit = topTable(fit2,coef=2, n = Inf)
limma_result_fit$metab_symbol <- row.names(limma_result_fit)
```

```r
name_map_emci = read.csv('extdata/name_map_metab.txt',row.names=1)
```

```r
#### plot volcano

fold_cutoff = 0.3

limma_result_fit$diffexpressed <- "NO"
# if log2Foldchange > 0.5 and pvalue < 0.05, set as "UP" 
limma_result_fit$diffexpressed[limma_result_fit$logFC > fold_cutoff & limma_result_fit$adj.P.Val < 0.05] <- "UP"
# if log2Foldchange < -0.5 and pvalue < 0.05, set as "DOWN"
limma_result_fit$diffexpressed[limma_result_fit$logFC < -fold_cutoff & limma_result_fit$adj.P.Val < 0.05] <- "DOWN"

limma_result_fit$delabel <- NA

for (i in 1:nrow(limma_result_fit)){
    if (limma_result_fit$diffexpressed[i] != "NO"){
        hmdb_name=limma_result_fit$metab_symbol[i]
        if (hmdb_name %in% name_map$HMDB){
            idx=which(name_map$HMDB==hmdb_name)
            real_name=name_map$Match[idx]
            limma_result_fit$delabel[i]=real_name
        }
    }
}
name_map$Query <- sub("^(.{4})(.{2})", "\\1", name_map$Query)

# Step 2: Set the modified Query column as the row names
name_map_lmci <- name_map
rownames(name_map_lmci) <- name_map_lmci$Query
name_map_lmci$Query <- NULL  # Remove the Query column after setting it as row names

limma_result_fit[rownames(name_map_lmci),]$delabel = name_map_lmci$Match
```

```r
limma_DE = limma_result_fit[which(limma_result_fit$diffexpressed != "NO"), ]
g<-ggplot(data=limma_result_fit, aes(x=logFC, y=-log10(adj.P.Val),col=diffexpressed)) +
        geom_point(size=2) +
        theme_minimal() +
        scale_color_manual(breaks=c("DOWN","NO","UP"),values=c("blue", "grey", "red")) +
        guides(color=guide_legend(title = "DE",override.aes = list(size = 3))) +
        geom_vline(xintercept=c(-fold_cutoff, fold_cutoff), col="brown") +
        geom_hline(yintercept=-log10(0.05), col="brown") + 
        xlim(-3, 3) +ylim(-0.1,30) +
        theme(axis.title = element_text(size = 15),axis.text=element_text(size=15),
              legend.title=element_text(size=15),legend.text=element_text(size=15))+
        geom_text_repel(aes(label=delabel),size=5,max.overlaps=11,force=10)+
theme_bw()+ggtitle('After Confounder Adjustment - EMCI',subtitle = paste0('Number of EMCI DE Metabolites: ',dim(limma_DE)[1]))
g

ggsave("figs/Figure_2H.png",plot = g, dpi=600, device = "png")    
```
