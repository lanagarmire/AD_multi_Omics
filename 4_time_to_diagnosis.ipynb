```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


EMCI1 = mpatches.Patch(color='red', label='EMCI1')
EMCI2 = mpatches.Patch(color='#FFA600', label='EMCI2')
LMCI1 = mpatches.Patch(color='#0000FF', label='LMCI1')
LMCI2 = mpatches.Patch(color='black', label='LMCI2')

x1 = np.arange(start=0, stop=17)*6
x2 = np.arange(start=0, stop=17)*6
x3 = np.arange(start=0, stop=17)*6
x4 = np.arange(start=0, stop=17)*6

####snf
y1   = prop_EMCI_1[0:17]
y2   = prop_EMCI_2[0:17]
y3   = prop_LMCI_1[0:17]
y4   = prop_LMCI_2[0:17]

fig, ax1 = plt.subplots(figsize=(20,8))
ax1.plot(x1, y1, '--o', color='red', linewidth=2, markersize=10)
ax1.plot(x2, y2, '--o', color='#FFA600', linewidth=2, markersize=10)
ax1.plot(x3, y3, '--o', color='#0000FF', linewidth=2, markersize=10)
ax1.plot(x4, y4, '--o', color='black', linewidth=2, markersize=10)

ax1.tick_params(axis='both', which='major', labelsize=20)
ax1.tick_params(axis='both', which='minor', labelsize= 15)


ax1.set_xlabel('months', fontsize = 20)
ax1.set_ylabel('Percentage of Dementia', color='black', fontsize = 20)
ax1.set_yticks(np.arange(min(y4), max(y4), 0.05), fontsize=14)
ax1.set_xticks(ticks=x3, fontsize=30)
ax1.grid(True)
plt.legend(handles=[EMCI1,EMCI2,LMCI1,LMCI2],fontsize=16,loc=0)
#plt.show()
plt.tight_layout()

plt.savefig('figs/Figure_4E.png', dpi=600)
```
