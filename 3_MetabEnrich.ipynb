```r
emci_pathway_result<-read.csv("DE_results/RP_emci_metabpathway.csv",header=T)
lmci_pathway_result<-read.csv("DE_results/RP_lmci_metabpathway.csv",header=T)
emci_pathway_result
lmci_pathway_result
```

```r
library(ggplot2)
library(ggrepel)
```

```r
ggplot(emci_pathway_result,aes(x=Impact,y=X.log10.p.,color=Raw.p,size=Impact))+
    geom_point()+
    theme_bw()+
    theme(panel.grid=element_blank(),axis.title = element_text(size = 12),axis.text=element_text(size=10),legend.title=element_text(size=12),legend.text=element_text(size=12))+
    coord_fixed(ratio=0.2)+
    scale_color_gradient(low="red",high="yellow")+
    labs(x="Pathway Impact",y="-log10(p value)",color="p value",size="Pathway Impact")+
    geom_text_repel(
    data = subset(emci_pathway_result, X.log10.p. >1.3),
    aes(label = X),
    size = 4,
    color="black",
    max.overlaps=5
  )
ggsave("figs/RP_3c.png",dpi=600,width=8,heigh=8)
```

```r
ggplot(lmci_pathway_result,aes(x=Impact,y=X.log10.p.,color=Raw.p,size=Impact))+
    geom_point()+
    theme_bw()+
    theme(panel.grid=element_blank(),axis.title = element_text(size = 12),axis.text=element_text(size=10),legend.title=element_text(size=12),legend.text=element_text(size=12))+
    coord_fixed(ratio=0.05)+
    scale_color_gradient(low="red",high="yellow")+
    labs(x="Pathway Impact",y="-log10(p value)",color="p value",size="Pathway Impact")+
    guides(color = guide_colorbar(order = 1),size = guide_legend(order = 2))+
    geom_text_repel(
    data = subset(lmci_pathway_result, X.log10.p. >2.5 ),
    aes(label = X),
    size = 4,
    color="black"
  )
ggsave("figs/RP_3f.png",dpi=600,width=8,heigh=8)
```
