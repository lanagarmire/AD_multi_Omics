```r
library(dplyr)
library(ggplot2)
library(ggpubr)

# Ensure grouping factor is ordered EMCI → LMCI
merge$labels_matching_UC_big <- factor(
  merge$labels_matching_UC_big,
  levels = c("EMCI", "LMCI")
)

# Function to build Wilcox p-value label
wilcox_label <- function(df, yvar) {
  pval <- wilcox.test(df[[yvar]] ~ df$labels_matching_UC_big)$p.value
  label <- ifelse(pval < 0.001,
                  "p < 0.001",
                  paste0("p = ", round(pval, 3)))
  return(label)
}

plot_marker <- function(df, yvar, ylab) {
  
  p_label <- wilcox_label(df, yvar)
  
  ggplot(df, aes(x = labels_matching_UC_big, 
                 y = .data[[yvar]], 
                 color = labels_matching_UC_big)) +
    geom_boxplot(outlier.shape = NA, fill = NA, size = 0.3) +
    geom_jitter(size = 1, width = 0.2, alpha = 0.7) +
    
    # ★ Use default palette but rename legend
    labs(color = "pred_label") +
    
    theme_bw(base_size = 15) +
    theme(
      panel.border = element_rect(color = "black", fill = NA, size = 1),
      legend.position = "none",     # ★ allow legend to show
      axis.text  = element_text(size = 15),
      axis.title = element_text(size = 15)
    ) +
    
    labs(x = "", y = ylab) +
    
    annotate(
      "text", 
      x = 1.5,
      y = max(df[[yvar]], na.rm = TRUE) * 1.05,
      label = p_label,
      size = 4
    )
}


p_ratio  <- plot_marker(merge, "P_AB42AB40ratio", "Aβ42/Aβ40 Ratio")
p_pt181  <- plot_marker(merge, "P_PT181V2",        "P-tau181")
p_pt217  <- plot_marker(merge, "P_PT217",          "P-tau217")
p_ab40   <- plot_marker(merge, "P_AB40",           "Aβ40")

p_ratio
p_pt181
p_pt217
p_ab40

```
