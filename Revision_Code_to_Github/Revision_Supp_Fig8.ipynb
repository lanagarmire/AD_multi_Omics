```r
get_patient_viscode <- function(PTID, adnimerge) {
  df <- adnimerge[adnimerge$PTID == PTID, "VISCODE", drop = FALSE]
  df$VISCODE
}
get_followup_viscounts <- function(adnimerge, PTID_vec) {
  sapply(PTID_vec, function(id) {
    vis <- get_patient_viscode(id, adnimerge)
    length(unique(vis[!is.na(vis)]))
  })
}
get_viscode_distribution <- function(adnimerge, PTID_vec) {
  vis_all <- unlist(
    lapply(PTID_vec, function(id) get_patient_viscode(id, adnimerge))
  )
  table(vis_all)
}
clean_viscode <- function(x) {
  x <- tolower(x)
  x[x == "bl"] <- "0"

  # Remove "m" prefix → numeric
  as.numeric(sub("^m", "", x))
}


visdist_EMCI1 <- get_viscode_distribution(adnimerge, EMCI_1_PTID$x)
visdist_EMCI2 <- get_viscode_distribution(adnimerge, EMCI_2_PTID$x)
visdist_LMCI1 <- get_viscode_distribution(adnimerge, LMCI_1_PTID$x)
visdist_LMCI2 <- get_viscode_distribution(adnimerge, LMCI_2_PTID$x)

library(dplyr)
library(tidyr)
library(ggplot2)

make_df <- function(vis_table, group_name) {
  df <- as.data.frame(vis_table)
  df$VISCODE <- df$vis_all
  df$Month <- clean_viscode(df$VISCODE)

  df %>%
    rename(Count = Freq) %>%
    mutate(Group = group_name)
}

df_all <- bind_rows(
  make_df(visdist_EMCI1,"EMCI1"),
  make_df(visdist_EMCI2,"EMCI2"),
  make_df(visdist_LMCI1,"LMCI1"),
  make_df(visdist_LMCI2,"LMCI2")
)
df_all[1:5,]
```

```r
get_modified_dementia_df <- function(i, adnimerge, PTID) {
  library(dplyr)
  library(tidyr)

  # subset for this individual
  df <- adnimerge[adnimerge$PTID == PTID[i],
                  c("PTID","EXAMDATE", "DX_bl","DX")]

  # ensure EXAMDATE is Date
  df$EXAMDATE <- as.Date(df$EXAMDATE)  # add format="..." if needed

  # remove rows with missing EXAMDATE
  df <- df[!is.na(df$EXAMDATE), ]

  # if no valid rows, return empty
  if (nrow(df) == 0) {
    return(data.frame(PTID = PTID[i],
                      timestamp = integer(0),
                      DX = character(0),
                      stringsAsFactors = FALSE))
  }

  # order by date
  df <- df[order(df$EXAMDATE), ]

  # fill DX
  df <- df %>% fill(DX, .direction = "updown")

  # define time range
  first_time <- min(df$EXAMDATE)
  last_time  <- max(df$EXAMDATE)
  diff_days  <- as.numeric(last_time - first_time)

  # if diff_days < 0 or very small, just set 1 interval
  if (is.na(diff_days) || diff_days <= 0) {
    time_interval <- 1
  } else {
    time_interval <- as.integer(diff_days / 182.5) + 1
  }

  # create output
  modified_df <- data.frame(
    PTID      = df$PTID[1],
    timestamp = seq_len(time_interval),
    DX        = NA_character_,
    stringsAsFactors = FALSE
  )

  k <- 1
  current_time <- first_time

  for (j in seq_len(time_interval)) {
    # move to later visit if more than ~6 months apart and not at last row
    while (k < nrow(df) && as.numeric(current_time - df$EXAMDATE[k]) >= 182.5) {
      k <- k + 1
    }
    modified_df$DX[j] <- as.character(df$DX[k])
    current_time <- current_time + 182.5
  }

  return(modified_df)
}
```

```r
# get the AD prop for a list

get_ad_prop <- function(adnimerge, PTID){
    merged_df = get_modified_dementia_df(1, adnimerge, PTID)[,c(-1)]

    for(i in 2:length(PTID)){
        df2 = get_modified_dementia_df(i, adnimerge, PTID)[,c(-1)]
        merged_df = merge(merged_df, df2, by = "timestamp", all.x = TRUE, all.y = TRUE)
    }
    rownames(merged_df) = merged_df$timestamp
    merged_df = merged_df[,c(-1)]
    merged_df[is.na(merged_df)] <- "No record"
    prop_list= list()
    for(i in 1:nrow(merged_df))
    {
        sum = 0
        for(j in 1: ncol(merged_df))
        {
            
            if(merged_df[i , j] == "MCI")
            {
                sum = sum + 1
            }
        }
        sum_dimentia = 0
         for(j in 1: ncol(merged_df))
        {
            
            if(merged_df[i , j] == "Dementia")
            {
                sum_dimentia = sum_dimentia + 1
            }
        }
        prop = sum/(sum+ sum_dimentia)
        prop_list[i] = 1 - prop
    }
    return(prop_list)
    
}
```

```r
get_modified_dementia_df <- function(i, adnimerge, PTID) {
    library(dplyr)
    library(tidyr)

    ptid_val <- PTID[i]
    
    # 1. Prepare and filter the data for the subject
    df <- adnimerge[adnimerge$PTID == ptid_val, 
                    c("PTID", "EXAMDATE", "DX_bl", "DX")] %>%
        mutate(EXAMDATE = as.Date(EXAMDATE)) %>%
        filter(!is.na(EXAMDATE) & !is.na(DX)) %>%
        arrange(EXAMDATE)

    if (nrow(df) == 0) {
        return(data.frame(PTID = ptid_val, timestamp = integer(0), DX = character(0)))
    }

    first_time <- min(df$EXAMDATE)
    last_time  <- max(df$EXAMDATE)
    diff_days  <- as.numeric(last_time - first_time)
    
    interval_days <- 182.5 # ~ 6 months
    
    # Calculate the number of full 6-month intervals
    time_interval <- floor(diff_days / interval_days) + 1
    
    # Create the sequence of target dates (182.5 day intervals)
    target_dates <- first_time + (0:(time_interval - 1)) * interval_days

    # 2. Create the target grid dataframe
    target_df <- data.frame(
        EXAMDATE = target_dates,
        timestamp = seq_len(time_interval)
    )
    
    # 3. Join and LOCF (Last Observation Carried Forward)
    # The crucial part: Merge target dates with recorded dates.
    # The 'roll=TRUE' functionality is not standard in base R/tidyverse joins, 
    # so we use a sequence of steps to emulate LOCF.
    
    # Step A: Combine recorded dates with target dates and keep diagnosis
    locf_df <- target_df %>%
        # Add the actual dates and their diagnoses
        full_join(df %>% select(EXAMDATE, DX), by = "EXAMDATE") %>%
        arrange(EXAMDATE) %>%
        # Fill missing DXs DOWNWARDS (LOCF) for all dates, including target dates.
        fill(DX, .direction = "down") %>%
        # Filter back to only the target dates (our 182.5 day time points)
        filter(EXAMDATE %in% target_dates) %>%
        # Select final output columns
        select(timestamp, DX) %>%
        mutate(PTID = ptid_val)

    # 4. Final check and return
    # If the first target date is before the first recorded visit, DX might be NA
    # We should ensure the first DX is based on the first recorded visit if the first timestamp is close
    if(is.na(locf_df$DX[1]) && nrow(df) > 0) {
        locf_df$DX[1] <- as.character(df$DX[1])
    }
    
    # If any remaining NAs (shouldn't happen with proper fill but safe to check), use "No record"
    locf_df <- locf_df %>%
        mutate(DX = replace_na(DX, "No record"))
    
    return(locf_df %>% select(PTID, timestamp, DX))
}
# --- 2. Main Function: Calculate Accumulated Dementia Proportion (CORRECTED) ---

get_ad_prop <- function(adnimerge, PTID){
    # 1. Merge all subjects' modified longitudinal data
    
    # Start with the first subject
    merged_df <- get_modified_dementia_df(1, adnimerge, PTID) %>% 
        select(-PTID) %>%
        rename(!!PTID[1] := DX) # Use !! to unquote the name, making the PTID the column name

    # Loop through the rest of the subjects and merge
    for(i in 2:length(PTID)){
        df_new <- get_modified_dementia_df(i, adnimerge, PTID) %>% 
            select(-PTID) %>%
            rename(!!PTID[i] := DX)
        
        # Merge by 'timestamp'
        merged_df <- merge(merged_df, df_new, by = "timestamp", all.x = TRUE, all.y = TRUE)
    }
    
    # Extract the diagnosis columns (remove timestamp)
    dx_matrix <- merged_df %>% select(-timestamp)

    # Replace NA (from subjects with shorter follow-up) with "No record"
    dx_matrix[is.na(dx_matrix)] <- "No record"
    
    # 2. Calculate the cumulative proportion

    # CRITICAL FIX: Define the total number of subjects in the cohort (the denominator)
    N_total <- ncol(dx_matrix) 
    
    # Initialize list for the proportions
    prop_list <- list()

    for(i in 1:nrow(dx_matrix)) { # Loop through each timestamp
        
        # Count subjects who have progressed to "Dementia" at this time point i
        sum_dementia <- sum(dx_matrix[i, ] == "Dementia")
        
        # Calculate the accumulated dementia risk (Cumulative Incidence)
        prop_dementia <- sum_dementia / N_total
        
        prop_list[[i]] <- prop_dementia # Use [[ ]] for list assignment
    }
    
    # Convert list to numeric vector for plotting
    return(unlist(prop_list))
}
```
